# Pyplot b√°sico

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'm')
plt.show()
```

```python
plt.plot(x,y, 'yD:')
plt.show()
```

## Histogramas

```python
plt.hist(y)  #histograma de los valores de y
plt.show()
```

## Pastel

```python
plt.pie(y)
plt.show()
```

## Correlacion entre variables

```python
plt.scatter(x,y)  #correlacion entre la variable x e y
plt.show()
```

## Distribucion de los datos

```python
plt.boxplot(x)
plt.show()
```

# Subplot

```python
plt.subplot(1,2,1) # Grafico 1
plt.plot(x,y, 'r--')

plt.subplot(1,2,2)  # Gr√°fico 2
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--') # Plot 1
plt.plot(y,x, 'b:') #Plot 2

plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1) #Grafico 1
plt.plot(x,y,'r--') #Plot 1
plt.plot(y,x,'b:')  #Plot 2

plt.subplot(2,1,2) #Grafico 2
plt.pie(y)         #Plot 1
plt.show()
```

# M√©todo orientado a objetos

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y, 'b')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.17,0.55,0.4,0.3])

axes.plot(x,y, 'b')
axes2.plot(y,x, 'r:')
axes2.set_facecolor('grey')
plt.show()
```

# Subplots

```python
fig, axes = plt.subplots()
axes.plot(x,y)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
plt.show()
```

```python
fig, (axes1,axes2) = plt.subplots(nrows=1, ncols=2)
axes1.plot(x,y,'b')
axes2.plot(y,x,'r')
plt.show()
```

```python
fig, axes = plt.subplots(2,4)

#genera un trazo accediendo a las graficas a traves del indice de la matriz
axes[0,0].plot(x,y)
axes[0,1].plot(y,x, 'r')

fig.tight_layout() #mejora la visualizaci√≥n de los ejes de cada gr√°fico
```

```python
fig, ((axes1, axes2, axes3, axes4), (axes5, axes6, axes7, axes8)) = plt.subplots(2,4)
axes1.plot(x, np.sin(x), 'b')
axes2.plot(y, np.cos(x), 'purple')
axes3.plot(x, np.tan(y), 'orange')
axes4.plot(y, np.cos(y)**2,'black')

#mejorar la visualizaci√≥n de las figuras
fig.tight_layout()
```

# Leyendas, etiquetas, t√≠tulos, tama√±o

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax2.plot(y,x, 'r')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b')
ax1.set_title("Relacion X - Y")

ax2.plot(y,x, 'r')
ax2.set_title("Relacion Y - X")
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', )
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

ax2.plot(y,x, 'r')
ax2.set_title("Relacion Y - X")
ax2.set_xlabel("Y")
ax2.set_ylabel("X")
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'b', label = "fx")
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend()
```

 # Bar Plot

```python
countrys = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", "GERMANY"]
population = [5000, 800, 900, 1000, 300]
```

```python
plt.bar(countrys,population)
plt.show()
```

```python
plt.bar(countrys,population,  width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.show()
```

```python
plt.bar(countrys,population, width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.xticks(np.arange(5), ('India','Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.show()
```

```python
plt.barh(countrys,population)
plt.show()
```

# Crear otro tipo de gr√°ficas

```python
data = np.random.randint(1,50,100)
```

```python
plt.hist(data)
plt.show()
```

```python
plt.hist(data, bins=50)
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step')
plt.show()
```

```python
# Muestra rango intercuartil y la mediana
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```

## Crear un gr√°fico tipo Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y, s=area)
plt.show()
```

```python
plt.scatter(x,y, s=area, c= colors, marker = 'v') # cambiar el marker
plt.show()
```

```python
# Transparencia
plt.scatter(x,y, s=area, c= colors, marker = 'o', alpha= 0.5)
plt.show()
```
